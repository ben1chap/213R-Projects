### Data Source 1: SQL

```python
#Creating SQL Connection Object and pulling into pandas
conn = sqlite3.connect('database.sqlite')
command = 'SELECT * FROM BoardGames'
games = pd.read_sql(command, conn)
games.head()
```

After pulling in the data set an interesting question could be: "What number of people should you invite over to allow the broadest access to board games in this dataset?" We will use the columns maxplayers and minplayers to create a visualization to answer this question.

I decided to drop the missing data since there were only three observations missing and it seemed to be missing randomly

```python
player = games[['game.id', 'details.maxplayers', 'details.minplayers']]
player.loc[player['details.minplayers'].isnull()]
player.dropna().head()
```

I now make a for loop to fill in Boolean values for whether the specified number of players could play the game based on the data set.

```python
for num_players in range(1, 101):
    column_name = 'num_players' + str(num_players)
    player[column_name] = (player['details.minplayers'] <= num_players) & (player['details.maxplayers'] >= num_players) 
player.head()
```

This is an error from pandas that doesn't apply to what I am working on

```python
counts = player.filter(like='num_players').sum().to_frame()
counts = counts.rename(columns={0: 'Number of Games'}).reset_index(drop=True)
counts['Number of Players'] = range(0, 100)
counts.head()
y = counts['Number of Games'].values
x = counts['Number of Players'].values
```

```python
text_kwargs = dict(fontsize=14, family='Arial')

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 4))
ax.bar(x, y, width=.95, color='.7', label='Average', edgecolor='black')
ax.set_ylabel('Number of Games', **text_kwargs)
ax.set_xlabel('Number of Players', **text_kwargs)
ax.legend(loc='upper right')
fig.suptitle('Optimal Number of Players to Play the Most Games', fontsize=20, fontweight='bold')
ax.bar
```

# Data Source: API (*Do not run again-- token is not included*)

This data comes from BYU's Handshake server, which hosts data on all the recruiting events that occur through BYU.

```python
#pull in the token
with open('handshake_token.txt', 'r') as f:
    token = f.read()
    
params = {'page': 0,
         'per_page': 50,
         'sort_direction': 'desc',
         'sort_column': 'start_date'}
    
headers = {
        'Authorization': 'Token token="{}"'.format(token),
        'Content-Type': 'application/json',
        'Cache-Control': 'no-cache--url',
    }
url = 'https://app.joinhandshake.com/api/v1/events'

all_events = []

#This loop will gather all the data we want from the API and organize it into a dataframe
while True:
    r = requests.get(url, params=params, headers=headers)
    if r.status_code != 200:
        continue
    if (r.json())['success'] != True:
        continue
    events = (r.json())['events']
    if len(events) == 0:
        break
    for event in events:
        single_event = []
        single_event.append(event['id'])
        single_event.append(event['name'])
        single_event.append(event['status'])
        single_event.append(event['invite_only'])
        single_event.append(event['start_date'])
        single_event.append(event['event_type']['name'])
        all_events.append(single_event)
    params['page'] += 1
```

```python
events = pd.DataFrame(all_events, columns=['ID', 'Name', 'Status', 'Invite Only', 'Start Date', 'Event Type'])
events.head()
```

An interesting visualization might be how many events have been held throughout the past and on which date they were held on

```python
events['Event Type'].value_counts()
```

```python
events['Status'].value_counts()
```

```python
events['Invite Only'].value_counts()
```

```python
#This narrows down the data frame to just the events we want to look at
events = events[(events['Event Type'] != 'Group Appointment') & (events['Event Type'] != 'Other')]
events = events[events['Status'] == 'approved']
events = events[events['Invite Only'] == False]
events.head()
```

```python
#This standardizes the time stamp placed on each event so they can be sorted into their correct dates
events['Start Date'] = events['Start Date'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S.%fZ'))
events['Start Date'] = events['Start Date'].dt.tz_localize('UTC')
events['Start Date'] = events['Start Date'].dt.tz_convert(pytz.timezone('US/Mountain'))
events['Start Date'] = events['Start Date'].apply(lambda x: x.date())
events['Start Date'].head()
```

```python
#This brings the desired data into two columns, the date and the number of events that day
info = events['Start Date'].value_counts().sort_index()
dt_index = pd.date_range(start=info.index[0], end=info.index[-1])
info.index = pd.DatetimeIndex(info.index)
info = info.reindex(dt_index, fill_value=0)
info.head(10)
```

```python
#Plots dates and number of events on each data via line graph
counts = info.values
dates = info.index.map(mdates.date2num).values

text_kwargs = dict(family='sans-serif', fontsize=12, fontstyle='italic')

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')
count_fmt = ticker.AutoMinorLocator(n=4)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(17, 4))
ax.plot(dates, counts)
ax.grid(which='both', color='.9')

ax.set_ylim(bottom=-.1, top=45)

ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)
ax.xaxis.set_minor_locator(months)
ax.yaxis.set_minor_locator(count_fmt)

ax.set_xlabel('Date', **text_kwargs)
ax.set_ylabel('Event Count', **text_kwargs)
fig.suptitle('Events Over Time', fontsize=20, fontweight='bold')

fig;
```

It appears that the number of events cycles through low and high periods based on the fall and winter semesters of the college school year.
